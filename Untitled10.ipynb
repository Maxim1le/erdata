```python
conditions = [df['FTHG'] > df['FTAG'], 
              df['FTHG'] < df['FTAG']]
choices = [df['HomeTeam'], df['AwayTeam']]
df['Win'] = np.select(conditions, choices, default='Tie')
df
```

```python
wine =pd.DataFrame(df.Win.value_counts())
wine.drop(['Tie'],inplace =True)
wine.index.name ='Team'
wine.Win.name = 'Wins'
```

```python
Home = df.groupby('HomeTeam')[['FTHG']].sum()
Away = df.groupby('AwayTeam')[['FTAG']].sum()
```

```python
HomeTeam = [] 
AwayTeam = []
idd = []
scoreH = []
scoreA = [] 
attendance = []
date = []
for i in r['attendances']['matches']:    
    scoreH.append(i['home-team']['score'])
    HomeTeam.append(i['home-team']['name'])
    scoreA.append(i['away-team']['score'])
    AwayTeam.append(i['away-team']['name'])
    date.append(i['date'])
    idd.append(i['id'])
    attendance.append(i['attendance'])
df5 = pd.DataFrame(data = {'date':date, 'id':idd, 'HomeTeam':HomeTeam, 'AwayTeam':AwayTeam, 'FTHG':scoreH, 'FTAG':scoreA, 'attendance':attendance })
df5.at[29, 'attendance'] = 60015
df5.at[40, 'attendance'] = 59959
df5 = df5.sort_values(by='HomeTeam', ignore_index=True)
#Поскольку на каждом стадионе есть гостевые болельщики, то, если Арсенал был гостевой командой, 
#умножим кол-во болельщиков на 0.25, а если домашней, то на 0.75\
df6 = df5.iloc[:23,:]
df6['attendance']=df6['attendance']*0.75
df7 = df5.iloc[23:,:]
df7['attendance']=df7['attendance']*0.25
frames = [df6, df7]
df8 = pd.concat(frames)
```

```python
df8["Goals scored"] = np.where(df8.HomeTeam =='Arsenal', df8.FTHG, df8.FTAG)
df8["Goals missed"] = np.where(df8.AwayTeam !='Arsenal', df8.FTAG, df8.FTHG)
df9=df8.iloc[:,6:]
df8
```
